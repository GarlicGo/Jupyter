```python
#!/usr/bin/env python
# -*- coding: UTF-8 -*-
"""
@File    ：PressureLocationProcess.py
@Author  ：zhaolin
@Date    ：2022/3/31 14:35
"""
from math import tan

import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

def cv_show(img):
    cv2.imshow('cv_show', img)
    cv2.waitKey(0)  # 等待时间，毫秒级，0标示按任意键终止
    cv2.destroyAllWindows()
    cv2.waitKey(1)  # 解决mac中窗口不关闭的问题
    
def showImages(images, titles = [''], size = [1, 1], BGR = False):
    for i in range(len(images)):
        plt.subplot(size[0], size[1], i + 1)
        showImage = images[i]
        if (BGR):
            b,g,r = cv2.split(images[i])
            showImage = cv2.merge([r,g,b])
        plt.imshow(showImage ,'gray')
        plt.title(titles[i])
        plt.xticks([]), plt.yticks([])

# 裁剪黑边
def cutVerticalBlack(img):
    new_img = []
    for i in range(0, len(img)):
        sizes = int(sum(img[i]) / 255)
        if sizes > 20:
            new_img.append(img[i])
    return new_img

def getChart(arr):
    fig, ax = plt.subplots(figsize=(10, 7))
    ax.bar(x=list(map(str, range(0, len(arr)))), height=arr)
    ax.set_title("Simple Bar Plot", fontsize=15)
```

```python
base = cv2.imread(
    '/Users/zhouxinyu/Code/Python/jupyter/Project/ZF/压线筒污染/windowImg/base.png', cv2.IMREAD_COLOR)

showImages([base], BGR=True)
```
